```python
def make_id(prefix: str, n: int, width: int = 3) -> str:
    return f"{prefix}{str(n).zfill(width)}"


def random_name() -> str:
    return f"{random.choice(first_names)} {random.choice(last_names)}"


def sample_skills(required_skills, preferred_skills, min_total=5, max_total=8):
    skills = set()

    # Most candidates should have decent overlap with their preferred category
    req_count = random.randint(max(2, len(required_skills) // 2), len(required_skills))
    pref_count = random.randint(1, min(3, len(preferred_skills)))

    skills.update(random.sample(required_skills, req_count))
    skills.update(random.sample(preferred_skills, pref_count))

    all_skills = list(skills_df["skill_name"])
    target_count = random.randint(min_total, max_total)

    while len(skills) < target_count:
        skills.add(random.choice(all_skills))

    return sorted(skills)


def compute_match_score_for_generation(candidate_skills, job_required_skills, exp_years, min_exp, education, required_edu,
                                       candidate_location, job_location, desired_category, job_category):
    candidate_set = set(s.lower() for s in candidate_skills)
    required_set = set(s.lower() for s in job_required_skills)

    skill_score = len(candidate_set & required_set) / len(required_set) if required_set else 0
    exp_score = 1.0 if exp_years >= min_exp else round(exp_years / min_exp, 2) if min_exp > 0 else 1.0

    edu_rank = {"High School": 1, "Diploma": 2, "Bachelor": 3, "Master": 4, "PhD": 5}
    edu_score = 1.0 if edu_rank.get(education, 0) >= edu_rank.get(required_edu, 0) else 0.0

    loc_score = 1.0 if candidate_location == job_location else 0.5
    pref_score = 1.0 if desired_category == job_category else 0.4

    score = (0.5 * skill_score + 0.2 * exp_score + 0.1 * edu_score + 0.1 * loc_score + 0.1 * pref_score) * 100
    return round(score, 1)


def weighted_status_from_score(score):
    if score >= 90:
        return random.choices(
            ["Interviewed", "Placed", "Applied"],
            weights=[40, 35, 25],
            k=1
        )[0]
    if score >= 80:
        return random.choices(
            ["Applied", "Interviewed", "Matched", "Placed"],
            weights=[40, 25, 25, 10],
            k=1
        )[0]
    if score >= 70:
        return random.choices(
            ["Matched", "Applied", "Interviewed", "Rejected"],
            weights=[40, 35, 10, 15],
            k=1
        )[0]
    return random.choices(
        ["Matched", "Rejected", "Applied"],
        weights=[50, 35, 15],
        k=1
    )[0]
```

```python
num_jobs = 60
jobs = []

for i in range(1, num_jobs + 1):
    template = random.choice(job_templates)
    title = random.choice(template["titles"])
    category = template["category"]
    location = random.choice(locations)
    work_mode = random.choice(work_modes)
    min_exp = random.choice([1, 2, 3, 4, 5, 6])
    education_required = random.choice(["Bachelor", "Bachelor", "Bachelor", "Master"])

    required_count = random.randint(4, min(6, len(template["required_skills"])))
    preferred_count = random.randint(2, min(4, len(template["preferred_skills"])))

    required_skills = random.sample(template["required_skills"], required_count)
    preferred_skills = random.sample(template["preferred_skills"], preferred_count)

    jobs.append({
        "job_id": make_id("J", i),
        "title": title,
        "category": category,
        "location": location,
        "work_mode": work_mode,
        "min_experience": min_exp,
        "education_required": education_required,
        "required_skills": ",".join(required_skills),
        "preferred_skills": ",".join(preferred_skills),
        "description": template["description"],
    })

jobs_df = pd.DataFrame(jobs)
jobs_df.head()
```

```python
num_candidates = 250
candidates = []

for i in range(1, num_candidates + 1):
    preferred_template = random.choice(job_templates)
    desired_category = preferred_template["category"]

    years_experience = random.choice([0, 1, 2, 3, 4, 5, 6, 7, 8])
    education_level = random.choice(["Bachelor", "Bachelor", "Master"])
    location = random.choice(locations)

    skills = sample_skills(
        preferred_template["required_skills"],
        preferred_template["preferred_skills"],
        min_total=5,
        max_total=8
    )

    candidates.append({
        "candidate_id": make_id("C", i),
        "full_name": random_name(),
        "location": location,
        "education_level": education_level,
        "years_experience": years_experience,
        "desired_job_category": desired_category,
        "skills_list": ",".join(skills),
    })

candidates_df = pd.DataFrame(candidates)
candidates_df.head()
```

```python
courses = []

for i, course in enumerate(courses_master, start=1):
    course_name, provider, skill_covered, duration_weeks, difficulty, course_type = course
    courses.append({
        "course_id": make_id("CR", i),
        "course_name": course_name,
        "provider": provider,
        "skill_covered": skill_covered,
        "duration_weeks": duration_weeks,
        "difficulty": difficulty,
        "course_type": course_type,
    })

courses_df = pd.DataFrame(courses)
courses_df.head()
```

```python
applications = []
application_counter = 1
today = datetime(2026, 3, 10)

for _, candidate in candidates_df.iterrows():
    desired_jobs = jobs_df[jobs_df["category"] == candidate["desired_job_category"]].copy()

    # Also allow some adjacent-role applications so the data is not too clean
    adjacent_jobs = jobs_df[jobs_df["category"] != candidate["desired_job_category"]].sample(5, random_state=random.randint(1, 9999))
    candidate_job_pool = pd.concat([
        desired_jobs.sample(min(6, len(desired_jobs)), random_state=random.randint(1, 9999)),
        adjacent_jobs
    ]).drop_duplicates(subset=["job_id"])

    applied_jobs = candidate_job_pool.sample(random.randint(2, 4), random_state=random.randint(1, 9999))

    candidate_skills = [s.strip() for s in candidate["skills_list"].split(",")]

    for _, job in applied_jobs.iterrows():
        job_required_skills = [s.strip() for s in job["required_skills"].split(",")]

        match_score = compute_match_score_for_generation(
            candidate_skills=candidate_skills,
            job_required_skills=job_required_skills,
            exp_years=float(candidate["years_experience"]),
            min_exp=float(job["min_experience"]),
            education=candidate["education_level"],
            required_edu=job["education_required"],
            candidate_location=candidate["location"],
            job_location=job["location"],
            desired_category=candidate["desired_job_category"],
            job_category=job["category"],
        )

        status = weighted_status_from_score(match_score)
        status_date = today - timedelta(days=random.randint(1, 45))

        applications.append({
            "application_id": make_id("A", application_counter),
            "candidate_id": candidate["candidate_id"],
            "job_id": job["job_id"],
            "match_score": match_score,
            "application_status": status,
            "status_date": status_date.strftime("%Y-%m-%d"),
        })
        application_counter += 1

applications_df = pd.DataFrame(applications)
applications_df.head()
```
